```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import io

plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置-黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
sns.set(font='SimHei')  # 解决Seaborn中文显示问题
```

```python
# 读取数据
data = io.loadmat('./happydata.mat')
values = data['data']
use_values = pd.DataFrame(values)
index_names = ['Score','Economy','Family','Health','Freedom','Generosity','Trust','Residual']
```

```python
# Pearson 相关系数
# xiangguan = np.corrcoef(values.T)
xiangguan = use_values.corr(method='pearson')
plt.figure(num=1, dpi=400)
labels = index_names
H = sns.heatmap(data=xiangguan,
                xticklabels=labels, yticklabels=labels,
                cmap=plt.get_cmap('Greens'),
                annot=True, fmt=".4f",
                annot_kws={'size':8, 'weight':'normal', 'color':'black'})
H.set_title('Pearson 相关系数矩阵')
```

```python
# Kendall tau 相关系数
xiangguan = use_values.corr(method='kendall')
plt.figure(num=1, dpi=400)
labels = index_names
H = sns.heatmap(data=xiangguan,
                xticklabels=labels, yticklabels=labels,
                cmap=plt.get_cmap('Greens'),
                annot=True, fmt=".4f",
                annot_kws={'size':8, 'weight':'normal', 'color':'black'})
H.set_title('Kendall tau 相关系数矩阵')
```

```python
# Kendall tau 相关系数
xiangguan = use_values.corr(method='spearman')
plt.figure(num=1, dpi=400)
labels = index_names
H = sns.heatmap(data=xiangguan,
                xticklabels=labels, yticklabels=labels,
                cmap=plt.get_cmap('Greens'),
                annot=True, fmt=".4f",
                annot_kws={'size':8, 'weight':'normal', 'color':'black'})
H.set_title('Spearman 相关系数矩阵')
```
